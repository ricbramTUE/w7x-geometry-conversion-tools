### DAGMC file merging

With this script I merge different `.h5m` and `.vtk` files for different geometry components created using `STEP_to_DAGMC.ipynb`: 

- I merge the `.h5m` files using the MOAB tool `mbconvert`;
- I merge the `.vtk` files using the MOAB toll `mbconvert`;

```python
# Imports and paths for merging h5m and vtk files
from pathlib import Path
import subprocess
import shutil


# ----------------------------
# Paths
# ----------------------------
# Adjust this base_folder to your geometry root
base_folder = Path("/home/riccardo/neutronics/Geometry/Module1_DAGMC")
# base_folder = Path("/home/riccardo/neutronics/Geometry/Geometry_files/W7X_M1_GeomCheck/vtk_files")
h5m_folder = base_folder / "h5m_files"
vtk_folder = base_folder / "vtk_files"
# vtk_folder = base_folder / "stahl paneele"

dagmc_tools_path = Path("/home/riccardo/dagmc-tools/bin")


# Collect H5M and VTK files
h5m_files = sorted([p for p in h5m_folder.glob("*.h5m")])
vtk_files = sorted([p for p in vtk_folder.glob("*.vtk")])
```

Merge the single component.h5m files into one module1.h5m 

```python
###################################
# 1. Merge H5M files using mbconvert
###################################

print("\n===== COMBINE H5M FILES =====")

merged_h5m = base_folder / f"{base_folder.name}.h5m"


mbconvert_path = shutil.which("mbconvert")

inputs = [str(p) for p in h5m_files]

cmd = [mbconvert_path] + inputs + [str(merged_h5m)]
subprocess.run(cmd, check=True)
```

Convert h5m file to vtk for visual inspection only!

```python
merged_h5m = base_folder / f"{base_folder.name}.h5m"
inspect_h5m_as_vtk = base_folder / f"{base_folder.name}_h5m_as_vtk.vtk"

cmd = ["mbconvert", str(merged_h5m), str(inspect_h5m_as_vtk)]
subprocess.run(cmd, check=True)

print(f"Converted {merged_h5m.name} -> {inspect_h5m_as_vtk}")
```

Check for leaky volumes

```python
cmd = [
    str(dagmc_tools_path / "check_watertight"),
    str(merged_h5m)
]

subprocess.run(cmd)
```

Check for overlaps

```python
cmd = [
    str(dagmc_tools_path / "overlap_check"),
    str(merged_h5m)
]

subprocess.run(cmd)
```

Merge the single component.vtk files into one module1.vtk 

```python
###################################
# 2. Merge VTK files using mbconvert
###################################

print("\n===== COMBINE VTK FILES =====")

merged_vtk = base_folder / f"{base_folder.name}.vtk"


mbconvert_path = shutil.which("mbconvert")

inputs = [str(p) for p in vtk_files]

cmd = [mbconvert_path] + inputs + [str(merged_vtk)]
subprocess.run(cmd, check=True)
```
